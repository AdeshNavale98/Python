# Subplots()

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
a=np.arange(10)
```

```python
a
```

```python
b=a*2
```

```python
b
```

```python
plt.plot(a,b)
```

```python
plt.plot(a,b,'bo')
```

```python
fig,axes=plt.subplots()
axes.plot(a,b,'r')
axes.set_xlabel("Value of X")
axes.set_ylabel("Value of Y")
axes.set_title("Graph of X vs Y")
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
#fig,axes=plt.subplots(nrows=2,ncols=6)
```

```python
fig
```

```python
axes
```

```python
for i in axes:
    i.plot(a,b,'r')
    i.set_xlabel("value of A")
    i.set_ylabel("value of B")
    i.set_title("Graph of A Vs B")

fig.tight_layout()
fig
```

```python
fig.savefig("Tuesday.png")
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(a,a**3,label='a**3')
ax.plot(a,a**2,label='a**2')
ax.legend()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(a,a**3,label='a**3')
ax.plot(a,a**2,label='a**2')
ax.legend(loc=1)
```

```python
fig,axes=plt.subplots(figsize=(10,5))
axes.plot(b,b**2,'b+')
axes.plot(b,b**3,'r+')
axes.set_title('Title')
plt.show()
```

```python
fig.savefig("Tuesday1.png")
```

```python
fig,axes=plt.subplots()
axes.plot(a,a+1,'r*')
axes.plot(a,a+2,'b*')
axes.plot(a,a+3,'g*')
```
